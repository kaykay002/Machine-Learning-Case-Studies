# REGRESSION MODEL CASE STUDY - PREDICTING APP RATING

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Sony\Downloads\googleplaystore.csv")
data
```

```python
data.info()
```

```python
#null vales in each column
data.isna().sum()
```

```python
#dropping all the null values of the target column i.e. RATING and other columns as well
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.head(100)
```

```python
#data encoding

from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
encoded_cat=le.fit_transform(data['Category'])
data['Category']=encoded_cat
```

```python
data['Category'].nunique()
```

```python
data.head()
```

```python
data['Reviews']=data['Reviews'].astype(int)
```

```python
data.info()
```

```python
def change_size(size):
    if 'M' in size:
        x = size[:-1]
        x=float(x)*1000000
        return x
    elif 'K' in size:
        x = size[:-1]
        x=float(x)*1000
        return x
    else:
        return None
```

```python
#applying user defined fun to standardise the size column

data['Size'] = data['Size'].map(change_size)
```

```python
data.info()
```

```python
#got some null values for size "varies with model", so ffill to deal with such values

data['Size'].fillna(method = 'ffill', inplace=True)
```

```python
data.info()
```

```python
#cleaning "Install" column

data['Installs']=[int(i.replace(',','').strip('+')) for i in data['Installs']]
```

```python
#encoding type
data['Type']=le.fit_transform(data['Type'])
```

```python
#price col to float, handling $ symbol

data['Price']=[float(i.strip('$')) for i in data['Price']]
```

```python
data.info()
```

```python
#encoding content rating and genre columns
data['Content Rating']=le.fit_transform(data['Content Rating'])
data['Genres']=le.fit_transform(data['Genres'])
```

```python
data.info()
```

```python
#dropping Last Updated,Current Ver, Android Ver

data=data.drop(columns=['Last Updated','Current Ver', 'Android Ver'])
```

```python
data.info()
```

```python
app_names=data['App']
data=data.drop(columns=['App'])
```

```python
data.info()
```

### Split the data into train and test data

```python
from sklearn.model_selection import train_test_split
X = data.drop(columns=['Rating'])
y = data['Rating']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# List of regression models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor()
}

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    print(f"\n{name} Results:")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"R-squared (R2): {r2:.4f}")
    print("----------------------------------------------\n")
```
